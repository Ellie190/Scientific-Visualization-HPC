# Applying Textures

```python
import pyvista as pv
from pyvista import examples
import numpy as np
from matplotlib.cm import get_cmap
import time
```

```python
#image_file = examples.mapfile
#image_file = 'lava.jpg'
start_time = time.time()
text_input1 = pv.read_texture('ancient.jpg')
text_input2 = pv.read_texture('brick.jpg')
text_input3 = pv.read_texture('forest.jpg')
text_input4 = pv.read_texture('galaxy.jpg')
text_input5 = pv.read_texture('lava.jpg')
text_input6 = pv.read_texture('ocean.jpg')
text_input7 = pv.read_texture('plant.jpg')
text_input8 = pv.read_texture('rock.jpg')
text_input9 = pv.read_texture('SB.jpg')
text_input10 = pv.read_texture('SW.jpg')
text_input11 = pv.read_texture('sun.jpg')
text_input12 = pv.read_texture('sand.jpg')
elapsed_time = time.time() - start_time
print(f"Elapsed time to computation "
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
text1 = examples.download_bird_texture()
text2 = examples.download_cake_easy_texture()
text3 = examples.download_emoji_texture()
text4 = examples.download_gourds_texture()
text5 = examples.download_masonry_texture()
text6 = examples.download_puppy_texture()
text7 = examples.download_rgba_texture()
text8 = examples.download_sky_box_nz_texture()
text9 = examples.download_usa_texture()
text10 = examples.load_globe_texture()
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation: "
      f"{elapsed_time:.3f} seconds")
```

```python
# create a structured surface
start_time = time.time()
x = np.arange(-200, 100, 25)
y = np.arange(-200, 100, 25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r)
curvsurf = pv.StructuredGrid(x, y, z)
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation: "
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
curvsurf.plot()
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
curvsurf.texture_map_to_plane(inplace=True)
curvsurf.plot(texture=text_input2)
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
# create a structured surface
x = np.arange(-10, 10, 0.25)
y = np.arange(-10, 10, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r)
skewsurf = pv.StructuredGrid(x, y, z)
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
skewsurf.plot()
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
# Map the curved surface to a plane - use best fitting plane
skewsurf.texture_map_to_plane(inplace=True)
skewsurf.plot(texture=text_input11)
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
sphere = pv.Sphere()
sphere.plot()
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
sphere.texture_map_to_sphere(inplace=True)# or text10
sphere.plot(texture=text10) #not 9 and 10
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
cy_surf = pv.Cylinder()
cy_surf.plot()
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```

```python
start_time = time.time()
cy_surf.plot(texture=text_input3) # not 9 and 10
elapsed_time = time.time() - start_time
print(f"Elapsed time for computation:"
      f"{elapsed_time:.3f} seconds")
```
